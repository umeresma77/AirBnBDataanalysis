```python
import pandas as pd

# Load the data
airbnb = pd.read_csv('AB_NYC_2019.csv')

# Check the first few rows
print(airbnb.head())
```

```python
print(airbnb.info())
```

```python
print(airbnb.describe())
```

```python
print(airbnb.isnull().sum())
```

```python
# Drop the unnecessary 'last_review' and 'reviews_per_month' columns
airbnb.drop(['last_review', 'reviews_per_month'], axis=1, inplace=True)
```

```python
# Replace missing 'name' and 'host_name' values with 'Unknown'
airbnb[['name', 'host_name']] = airbnb[['name', 'host_name']].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of prices
sns.histplot(airbnb['price'], bins=50)
plt.title('Distribution of Prices')
plt.show()
```

```python
# Number of listings in each neighborhood group
sns.countplot(x='neighbourhood_group', data=airbnb)
plt.title('Number of Listings in Each Neighborhood Group')
plt.show()
```

```python
# Average price in each neighborhood group
average_prices = airbnb.groupby('neighbourhood_group')['price'].mean()
print(average_prices)
```
